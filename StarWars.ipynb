```python
# Imports:
import pandas as pd
import numpy as np
```

```python
# Read in the dataset:
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
star_wars.head(10)
```

Will need to remove any rows where 'RespondentID' is NaN.

# Cleaning the dataset

```python
star_wars.dropna(inplace=True, subset=['RespondentID'])
star_wars.head(10)
```

```python
yes_no_mapping = {'Yes': True, 'No': False}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no_mapping)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no_mapping)
```

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].head()
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].head()
```

```python
seen_or_nan = {
    'Star Wars: Episode I  The Phantom Menace': True,
    'Star Wars: Episode II  Attack of the Clones': True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True,
    np.NaN: False,
}
```

```python
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(seen_or_nan)
```

```python
star_wars.head()
```

```python
columns = {}
for i, col in enumerate(star_wars.columns[3:9]):
    columns[col] = 'seen_' + str(i + 1)

star_wars = star_wars.rename(columns=columns)
```

```python
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars.head()
```

```python
columns = {}
for i, col in enumerate(star_wars.columns[9:15]):
    columns[col] = 'ranking_' + str(i + 1)
    
star_wars = star_wars.rename(columns=columns)
star_wars.head()
```

# Visualizing the most popular Star Wars movie

```python
mean_rankings = star_wars.loc[:,'ranking_1':'ranking_6'].mean()
mean_rankings
```

```python
%matplotlib inline
```

```python
mean_rankings.plot.bar()
```

The most popular star wars movie among survey participants was 'Episode V The Empire Strikes Back', followed by 'Episode VI Return of the Jedi' , and then 'Episode IV A New Hope'. The rankings of these three movies far outweighed movies one, two, and three.

Perhaps more people have seen movies five, six, and four.

```python
star_wars.loc[:,'seen_1':'seen_6'].sum().plot.bar()
```

More people have seen movies 4, 5, and 6 compared with the earlier ones, excluding the first one. This proves my earlier hypothesis that more people had seen movies 4 to 6.

There appears to be a slight pattern occurring between the two bar graphs in which, on average, the most seen movies receive the highest rating, and the least seen movies receive the lowest rating.

```python
star_wars_fans = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] == True]
```

```python
star_wars_fans.loc[:,'ranking_1':'ranking_6'].mean().plot.bar()
```

```python
star_wars_fans.loc[:,'seen_1':'seen_6'].sum().plot.bar()
```

Clearly, a higher majority of those who considered themselves Star Wars fans, rated movies 5, 6, and 4 the highest. However, between the two graphs the same pattern is occurring, in which the most watched movies are ranked the highest.

```python
star_trek_fans = star_wars[star_wars['Do you consider yourself to be a fan of the Star Trek franchise?'] == 'Yes']
```

```python
star_trek_fans.loc[:,'ranking_1':'ranking_6'].mean().plot.bar()
```

```python
star_trek_fans.loc[:,'seen_1':'seen_6'].sum().plot.bar()
```

A higher percentage of Star Trek fans have seen the movies 5, 6, and 4 the most, however, like all the other groups, have rated the most watched movies, the highest.
